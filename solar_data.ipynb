```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv('PVSoilingTest.csv')
```

```python
voltage1 = df['BusVoltage 1 (V)'] 
current1 = df['Current 1 (mA)'] 
power1 = df['Power 1 (mW)'] 
resistance1 = df['Resistance 1 (Ohm)'] 

voltage2 = df['BusVoltage 2 (V)'] 
current2 = df['Current 2 (mA)'] 
power2 = df['Power 2(mW)'] 
resistance2 = df['Resistance 2 (Ohm)'] 

time = df.index
```

```python
fig, ((ax1, ax2),(ax3, ax4))= plt.subplots(2, 2)
fig.tight_layout()

ax1.plot(time, voltage1, 'b')
ax1.plot(time, voltage2, 'r')
ax1.set_ylabel('Voltage (V)')

ax2.plot(time, current1, 'b')
ax2.plot(time, current2, 'r')
ax2.set_ylabel('Current (mA)')

ax3.plot(time, power1, 'b')
ax3.plot(time, power2, 'r')
ax3.set_xlabel('Time (0.5 s)')
ax3.set_ylabel('Power (mW)')

ax4.plot(time, resistance1, 'b')
ax4.plot(time, resistance2, 'r')
ax4.set_xlabel('Time (0.5 s)')
ax4.set_ylabel('Resistance (Ohm)')

plt.show()
```

```python
y1 = df['Power 1 (mW)'] 
y2 = df['Power 2(mW)']
x = df.index

plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r')
plt.xlabel('Time (0.5 s)')
plt.ylabel('Power (mW)')
plt.show()
```
